```python
import numpy as np
import pandas as pd
```

```python
# Funciones de impureza según el libro de Raschka
def entropy(counts):
    # -Σ p * log2(p)
    total = sum(counts)
    if total == 0: return 0
    probs = [c / total for c in counts]
    return -sum([p * np.log2(p) for p in probs if p > 0])

def gini(counts):
    # 1 - Σ p^2
    total = sum(counts)
    if total == 0: return 0
    probs = [c / total for c in counts]
    return 1 - sum([p**2 for p in probs])

def error(counts):
    # 1 - max(p)
    total = sum(counts)
    if total == 0: return 0
    probs = [c / total for c in counts]
    return 1 - max(probs)
```

```python
# Funcion para calcular la Ganancia de Información (IG)
def information_gain(parent, left, right, impurity_fn):
    n_p, n_l, n_r = sum(parent), sum(left), sum(right)
    ig = impurity_fn(parent) - (n_l / n_p) * impurity_fn(left) - (n_r / n_p) * impurity_fn(right)
    return ig
```

```python
# Datos del ejercicio 2
parent = (30, 100)
esc_a = {"left": (10, 50), "right": (20, 50)}
esc_b = {"left": (15, 40), "right": (15, 60)}
```

```python
# Calcular y comparar
results = []
measures = [("Entropía (Ih)", entropy), ("Gini (Ig)", gini), ("Error (Ie)", error)]

for name, fn in measures:
    ig_a = information_gain(parent, esc_a["left"], esc_a["right"], fn)
    ig_b = information_gain(parent, esc_b["left"], esc_b["right"], fn)
    results.append({
        "Medida": name,
        "IG Escenario A": round(ig_a, 4),
        "IG Escenario B": round(ig_b, 4),
        "Mejor Escenario": "A" if ig_a > ig_b else ("B" if ig_b > ig_a else "Empate")
    })

df_results = pd.DataFrame(results)
print(df_results)
```
